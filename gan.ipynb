GAN 개념

GAN은 데이터를 생성하는 모델을 만들고 싶을 때 사용합니다.

근데 그냥 예전처럼 모델하나만 만들지 말고 두개의 모델을 만드시면 됩니다.

Generator : 가짜 데이터를 생성하는 모델

Discriminator : 생성된 데이터가 진짜인지 가짜인지 구분하는 모델

이렇게 두개 만드시고 연결하면 그게 GAN입니다.

GAN은 Generative Adversarial Network 의 약자입니다.

Adversarial이 맞짱깐다는 소리인데..

왜 싸우는거냐면 GAN 모델 동작방식은 이렇기 때문입니다.

Generator는 데이터를 생성하고 (이게 우리가 필요한 핵심 모델입니다)

그 데이터를 진짠지 가짠지 Discriminator 모델이 검증하는 겁니다.

그래서 GAN의 핵심은 Generator 모델 vs Discriminator 모델 이렇게 둘이 경쟁시키는겁니다.

이 두개를 동시에 트레이닝해서 Discriminator가 더이상 진짜사진과 가짜사진을 구분할 수 없으면

Generator가 뛰어난 성능을 가질 수 있겠죠?

더 자세히 각각 모델의 역할을 이야기하려면 일단 프로젝트를 진행하며 설명드리겠습니다.


GAN 프로젝트 : 사람얼굴 생성기

일단 "사람 얼굴" 이미지를 생성하는 딥러닝 모델을 만들고 싶다고 가정해봅시다.

모델에다가 랜덤한 숫자를 입력하면 지가 알아서 사람 얼굴사진을 생성해주는겁니다.

그 사람 얼굴은 기계가 생성했으니 전에 없던 새로운 사람이겠죠.

GAN으로 만들어봅시다. 그럼 모델 두개를 만들어야한다고 했죠?

각각의 모델은 어떤 역할이 있는지 구체적으로 정리해드리겠습니다.

Generator

- 무작위 숫자를 입력하면 64x64x3 사이즈의 사람얼굴 이미지를 생성함

(여기서의 이미지들은 전부 숫자화된 이미지를 뜻합니다)

- 생성한 데이터를 discriminator에 전해줍니다.


Discriminator

- 64x64x3 사람 이미지를 입력해서 이 이미지가 진짜인지 가짜인지 판단하는 모델임

(마지막 레이어는 Dense(1, 활성함수="sigmoid") 이런거겠네요)

- generator에서 생성한 가짜 이미지도 집어넣을 수 있고 진짜 사람 이미지도 집어넣을 수 있음


각 모델의 역할은 이게 전부입니다. 이거 두개를 접합하면 GAN이 되는 것이고요.

그림으로 그리자면 이렇습니다.


▲ Random noise는 정말 랜덤한 숫자입니다. 미국놈들은 noise라고 많이 부릅니다.

참고로 여기서 Conv 레이어를 적용하면 DCGAN 이라고 부릅니다.

모델의 형태에 따라 CycleGAN, StarGAN, ProGAN 등 여러가지 종류가 있습니다.


GAN 트레이닝하는 법

이런 모델은 복잡해서 model.fit() 이걸로 쉽게 트레이닝할 수 없습니다.

트레이닝 과정을 직접 for 반복문으로 코드를 짜셔야하는데

어떤 식인지 설명해드리자면 일단 GAN 모델은 generator -> discriminator 순으로 연결되어있으니

마지막 레이어와 가까운 discriminator 모델부터 트레이닝하고 generator를 트레이닝합니다.

이걸 원하는 만큼 반복해주면 됩니다.


그래서 님들이 짤 코드 :

1. 랜덤숫자를 generator에 집어넣어서 가짜 이미지 50장을 생성

2. 진짜 사람 이미지 50장을 데이터셋에서 가져옴

3. discriminator 모델에 train_on_batch() 함수를 써서 이미지 100장으로 살짝 트레이닝함

4. 랜덤숫자 noise를 1로 라벨링한 (진짜이미지로 라벨링) 데이터셋을 만들기

5. 그 데이터셋으로 GAN 전체를 train_on_batch() 함수를 써서 트레이닝하기 (이게 generator 트레이닝)

6. 원하는 만큼 반복

준비한 진짜 사람 이미지가 동날 때 까지 반복문을 돌리면 epoch 1회 트레이닝이 끝납니다.

이건 model.fit 처럼 epoch 정보를 출력해주지 않는 그냥 일반 for 반복문이기 때문에

트레이닝 중간중간에 뭔가 출력해보고 싶으면 print 알아서 잘 쓰시면 됩니다.

하지만 5번 이해가십니까. 왜 GAN 전체에 noise를 집어넣는데 이게 왜 generator 트레이닝임?

이런거 궁금해야 여러분 훌륭한 대학원생, 훌륭한 연구원이 될 수 있습니다.

5번이 왜 generator 모델 트레이닝이냐면


GAN 모델은 generator -> discriminator 이렇게 붙어있습니다.

GAN에 noise라는 랜덤숫자를 집어넣으면 일단 generator를 거치며 진짜 이미지가 됩니다.

그리고 이게 진짜 이미지인척 discriminator에 투입됩니다. (Y데이터는 1로 라벨링 했으니까 진짜 이미지인척 하는거임)

그럼 discriminator는 이걸 구분하는 과정을 거치게 되고 loss가 나오게 됩니다.

하지만 이번 데이터는 답안이 전부 1로 라벨링이 되어있기 때문에 전체 GAN 입장에서는 loss를 최소화하려면 어떻게 해야할까요?

generator가 더욱 더 진짜같은 가짜 이미지를 생성해야합니다.

그래서 이러한 loss가 최소화 되는 방향으로 전체 w값을 수정하게 되고 그래서 generator 학습이 일어납니다.



이러한 원리를 알면 응용도 가능하기 때문에 원리학습이 언제나 중요합니다.

하지만 머리가 굳어 이해가 안된다면 그냥 따라해보는 것도 좋은 학습방법입니다.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

경로들 = os.listdir('img_align_celeba/img_align_celeba')

images = []

for path in 경로들[0:50000]:
  숫자화된거 = Image.open('img_align_celeba/img_align_celeba/' + path).crop((20, 30, 160, 180)).convert('L').resize((64,64))
  images.append( np.array(숫자화된거) )

images = np.array(images)
images = np.divide(images, 255)

images = images.reshape( 50000,64,64,1 )
print(images.shape)
```

```python
import tensorflow as tf

discriminator = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(64, (3,3), strides=(2, 2), padding='same', input_shape=[64,64,1]),
  tf.keras.layers.LeakyReLU(alpha=0.2),
  tf.keras.layers.Dropout(0.4),
  tf.keras.layers.Conv2D(64, (3,3), strides=(2, 2), padding='same'),
  tf.keras.layers.LeakyReLU(alpha=0.2),
  tf.keras.layers.Dropout(0.4),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
generator = tf.keras.models.Sequential([
  tf.keras.layers.Dense(4 * 4 * 256, input_shape=(100,) ),
  tf.keras.layers.Reshape((4, 4, 256)),
  tf.keras.layers.Conv2DTranspose(256, 3, strides=2, padding='same'),
  tf.keras.layers.LeakyReLU(alpha=0.2),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2DTranspose(128, 3, strides=2, padding='same'),
  tf.keras.layers.LeakyReLU(alpha=0.2),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2DTranspose(64, 3, strides=2, padding='same'),
  tf.keras.layers.LeakyReLU(alpha=0.2),
  tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Conv2DTranspose(1, 3, strides=2, padding='same', activation='sigmoid')
])
```

```python
GAN = tf.keras.models.Sequential([generator, discriminator])

discriminator.compile(optimizer='adam', loss='binary_crossentropy')
discriminator.trainable = False

GAN.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
랜덤숫자 = np.random.uniform(-1, 1, size=(8, 100))
예측 = generator.predict(랜덤숫자)

# for i in range(10):
#   plt.subplot(2, 5, i+1)
#   plt.imshow(예측[i].reshape(64,64), cmap='gray')
#   plt.axis('off')
# plt.tight_layout()
# plt.show()
```

```python
from datetime import datetime


x데이터 = images

for j in range(300):
  now = datetime.now()
  print(f'지금 epoch 몇회냐면 {j}, {now.minute}분 { now.second}초')
  for i in range(50000//128):
    #discriminator 트레이닝
    if i % 100 == 0:
      now = datetime.now()
      print(f'지금 몇 번째 batch냐면 {i}, {now.minute}분 {now.second}초')
    진짜사진들 = x데이터[i*128:(i+1)*128]
    마킹한정답들1 = np.ones(shape=(128, 1))

    loss1 = discriminator.train_on_batch(진짜사진들, 마킹한정답들1)

    랜덤숫자 = np.random.uniform(-1, 1, size=(128, 100))
    가짜사진들 = generator.predict(랜덤숫자, verbose=0)
    마킹한정답들0 = np.zeros(shape=(128, 1))

    loss2 = discriminator.train_on_batch(가짜사진들, 마킹한정답들0)

    #generator 트레이닝
    랜덤숫자 = np.random.uniform(-1, 1, size=(128, 100))
    마킹한정답들1 = np.ones(shape=(128, 1))
    loss3 = GAN.train_on_batch(랜덤숫자, 마킹한정답들1)
  print(f'이번 epoch 최종 loss는 discriminator {loss1 + loss2} GAN {loss3}')
```
